## Import grant data

```python
grant_path = r"P3_GrantExport.csv"
grant_data = pd.read_csv(grant_path, sep=";")
```

There is a type issue with "Approved Amount" column, indeed some strings are included in the column so let's clean this

### Cleaning grant amounts

```python
test=[]
for amt_idx, amt in enumerate(grant_data["Approved Amount"]):
    try:
        test.append(float(amt))
    except:
        test.append(None)
```

```python
%matplotlib inline
df = pd.DataFrame(test)
```

```python
df[0] = pd.to_numeric(df[0])
```

```python
df.boxplot(return_type="axes")
```

```python
import matplotlib.pyplot as plt
```

```python
fig, ax = plt.subplots()
#series.hist(ax=ax, bins=100, bottom=0.1)
df.hist(ax=ax, bins=100, bottom=0.1)
ax.set_yscale('log')
ax.set_title('Histogram of Approved Grant amount (CHF)')
ax.set_xlabel('Grant amount (CHF)')
ax.set_ylabel('Amount of projects with such amount')
```
